```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import seaborn as sns

df = pd.read_csv('data1.csv', index_col='Unnamed: 0')
df = df.dropna()

data_germany = df[df['country'] == 'DE']
data_france = df[df['country'] == 'FR']
data_italy = df[df['country'] == 'IT']
data_sweden = df[df['country'] == 'SE']
data_uk = df[df['country'] == 'UK']
data_usa = df[df['country'] == 'US']

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

axs[0, 0].scatter(data_germany['age'], data_germany['tax_top1'], c="#7FFF00")
axs[0, 0].set_title('Германия')

axs[0, 1].scatter(data_france['age'], data_france['tax_top1'], c = '#AAFF32')
axs[0, 1].set_title('Франция')

axs[0, 2].scatter(data_italy['age'], data_italy['tax_top1'], c = '#BBF90F')
axs[0, 2].set_title('Италия')

axs[1, 0].scatter(data_sweden['age'], data_sweden['tax_top1'], c = '#76FF92')
axs[1, 0].set_title('Швеция')

axs[1, 1].scatter(data_uk['age'], data_uk['tax_top1'], c = '#C1F80A')
axs[1, 1].set_title('Великобритания')

axs[1, 2].scatter(data_usa['age'], data_usa['tax_top1'], c = '#9ACD32')
axs[1, 2].set_title('США')
fig.suptitle('Взгляды на ставку налога в зависимости от возраста отдельно по каждой стране(по оси x - возраст, по оси y - ставка)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import seaborn as sns

def gradientbars(bars):
    grad = np.atleast_2d(np.linspace(0,1,256)).T
    ax = bars[0].axes
    lim = ax.get_xlim()+ax.get_ylim()
    for bar in bars:
        bar.set_zorder(1)
        bar.set_facecolor("none")
        x,y = bar.get_xy()
        w, h = bar.get_width(), bar.get_height()
        ax.imshow(grad, extent=[x,x+w,y,y+h], aspect="auto", zorder=0)
    ax.axis(lim)

df = pd.read_csv('data1.csv', index_col='Unnamed: 0')

df_out = df.groupby(['party_voted','q_inequality_problem']).size().reset_index(name='count')
s = list(df_out['count'])
ind = np.arange(7)
width = 0.15
xvals = s[0::5]
bar1 = plt.bar(ind, xvals, width, color = 'r',label='Совсем не проблема')
yvals = s[1::5]
bar2 = plt.bar(ind+width, yvals, width, color='g',label='Небольшая проблема')
zvals = s[2::5]
bar3 = plt.bar(ind+width*2, zvals, width, color = 'b',label='Проблема')
avals = s[3::5]
bar4 = plt.bar(ind+width*3, avals, width, color = 'y',label='Серьезная проблема')
cvals = s[4::5]
bar5 = plt.bar(ind+width*4, cvals, width, color = 'orange',label='Очень серьезная проблема')

xtic = ['Не укл. в класс.', 'правая', 'правоцентр', 'центр', 'левоцентр','левая','другое']
plt.xticks(ind, xtic ,rotation ='vertical')
plt.xlabel("Политическая партия")
plt.ylabel("Количество людей")
plt.legend()
plt.show()

mn = ['Совсем не проблема','Небольшая проблема','Проблема','Серьёзная проблема','Очень серьёзная проблема']


plt.subplot(3, 3, 1)
xvals1 = s[0:5]
b1 = plt.bar(range(1,6), xvals1)
plt.xticks(range(1,6), fontsize=10)
plt.title('Не укл. в класс.')
plt.xlabel('Отн. к нер. по дох.')
plt.ylabel('Кол-во чел, разд.мнение')

plt.subplot(3, 3, 2)
yvals1 = s[5:10]
b2 = plt.bar(range(1,6), yvals1)
plt.xticks(range(1,6), fontsize=10)
plt.title('правая')
plt.xlabel('Отн. к нер. по дох.')


plt.subplot(3, 3, 3)
xvals1 = s[10:15]
b3 = plt.bar(range(1,6), xvals1)
plt.xticks(range(1,6), fontsize=10)
plt.title('правоцентр')
plt.xlabel('Отн. к нер. по дох.')
plt.ylabel('Кол-во чел, разд.мнение')

plt.subplot(3, 3, 4)
yvals1 = s[15:20]
b4 = plt.bar(range(1,6), yvals1)
plt.xticks(range(1,6), fontsize=10)
plt.title('центрист')
plt.xlabel('Отн. к нер. по дох.')


plt.subplot(3, 3, 5)
xvals1 = s[20:25]
b5 = plt.bar(range(1,6), xvals1)
plt.xticks(range(1,6), fontsize=10)
plt.title('левоцентр')
plt.xlabel('Отн. к нер. по дох.')
plt.ylabel('Кол-во чел, разд.мнение')

plt.subplot(3, 3, 6)
yvals1 = s[25:30]
b6 = plt.bar(range(1,6), yvals1)
plt.xticks(range(1,6), fontsize=10)
plt.title('левая')
plt.xlabel('Отн. к нер. по дох.')


plt.subplot(3, 3, 7)
yvals1 = s[30:35]
b7 = plt.bar(range(1,6), yvals1)
plt.xticks(range(1,6), fontsize=10)
plt.title('другие')
plt.xlabel('Отн. к нер. по дох.')
plt.ylabel('Кол-во чел, разд.мнение')


gradientbars(b1)
gradientbars(b2)
gradientbars(b3)
gradientbars(b4)
gradientbars(b5)
gradientbars(b6)
gradientbars(b7)
# space between the plots
plt.tight_layout()
plt.show()

#df1['count_p'] = df.groupby(['party_voted','q_inequality_problem']).agg({'respondent_ID':'count'})

print('\nЕсли важны численные значения, то лучше посмотреть на сводную таблицу:\n')
pd.pivot_table(df, values='respondent_ID', index=['party_voted'], columns=['q_inequality_problem'], aggfunc='count')
```
